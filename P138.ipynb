```python
import pandas as pd
df1 = pd.read_csv('shared_articles.csv')
df1.head()
```

```python
df2 = pd.read_csv('users_interactions.csv')
df2.head()
```

```python
df1 = df1 [df1['eventType']=='CONTENT SHARED']
df1.head()
```

```python
def totalevent(df1_row):
  total_likes = df2[(df2['contentId']==df1_row['contentId'])&(df2['eventType']=='LIKE')].shape[0]
  total_follow = df2[(df2['contentId']==df1_row['contentId'])&(df2['eventType']=='FOLLOW')].shape[0]
  total_view = df2[(df2['contentId']==df1_row['contentId'])&(df2['eventType']=='VIEW')].shape[0]
  total_bookmark = df2[(df2['contentId']==df1_row['contentId'])&(df2['eventType']=='BOOKMARK')].shape[0]
  total_comment_created = df2[(df2['contentId']==df1_row['contentId'])&(df2['eventType']=='COMMENT CREATED')].shape[0]

  return total_likes + total_follow + total_view + total_bookmark + total_comment_created
df1 ['totalevent'] = df1.apply(totalevent, axis=1)
df1.head()
```

```python
df1 = df1.sort_values('totalevent',ascending = False)
df1.head()
```
